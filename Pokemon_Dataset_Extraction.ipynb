Iterating same method over all pages and appending it to form a dataset.

```python
rows = []
for i in range(1,1026):
  result = requests.get(f'https://pokeapi.co/api/v2/pokemon/{i}/').json()
  result2 =  requests.get(f'https://pokeapi.co/api/v2/pokemon-species/{i}/').json()
  id = result['id']
  name = result['name']
  height = result['height']
  weight = result['weight']
  stats_dict = { s['stat']['name']:s['base_stat'] for s in result['stats']}

  stats_hp = stats_dict['hp']
  stats_atk = stats_dict['attack']
  stats_def = stats_dict['defense']
  stats_spatk = stats_dict['special-attack']
  stats_spdef = stats_dict['special-defense']
  stats_spd = stats_dict['speed']
  types = result['types']
  if result['past_types']:
    past_type = result['past_types'][0]['types'][0]['type']['name']
  else:
    past_type = None
  if len(types) ==1:
    type1 = types[0]['type']['name']
    type2 = None
  elif len(types) == 2:
    type1 = types[0]['type']['name']
    type2 = types[1]['type']['name']
  gen = result2['generation']['name']
  capture_rate = result2['capture_rate']
  base_happiness = result2['base_happiness']
  baby = result2['is_baby']
  legendary = result2['is_legendary']
  mythical = result2['is_mythical']
  hatch_counter = result2['hatch_counter']
  if len(result2['egg_groups'])==1:
    egg_group1 = result2['egg_groups'][0]['name']
    egg_group2 = None
  elif len(result2['egg_groups'])==2:
    egg_group1 = result2['egg_groups'][0]['name']
    egg_group2 = result2['egg_groups'][1]['name']

  row = {
      'id':id,
      'Name':name,
      'Height':height,
      'Weight':weight,
      'HP':stats_hp,
      'Attack':stats_atk,
      'Defense':stats_def,
      'Sp.Atk':stats_spatk,
      'Sp.Def':stats_spdef,
      'Speed':stats_spd,
      'Type_1':type1,
      'Type_2':type2,
      'Is_Legendary':legendary,
      'Is_Mythical':mythical,
      'Egg_Group_1':egg_group1,
      'Egg_Group_2':egg_group2,
      'Generation':gen,
      'Capture_Rate':capture_rate,
      'Base_Happiness':base_happiness,
      'Is_Baby':baby,
      'Egg_Cycles':hatch_counter,
      'Past_Type':past_type
  }

  rows.append(row)
pokedata = pd.DataFrame(rows)
pokedata
```

```python
pokedata['Name'] = pokedata['Name'].str.capitalize()
pokedata['Type_1'] = pokedata['Type_1'].str.capitalize()
pokedata['Type_2'] = pokedata['Type_2'].str.capitalize()
pokedata['Egg_Group_1'] = pokedata['Egg_Group_1'].str.capitalize()
pokedata['Egg_Group_2'] = pokedata['Egg_Group_2'].str.capitalize()
pokedata['Generation'] = pokedata['Generation'].str.capitalize()
pokedata['Past_Type'] = pokedata['Past_Type'].str.capitalize()
pokedata
```

Basic Normalization

```python
def numerical(row):
  if row.endswith('-i'):
    row = row.rstrip('-i')
    row = row[0:3]+'-1'
    return row
  elif row.endswith('-ii'):
    row = row.rstrip('-ii')
    row = row[0:3]+'-2'
    return row
  elif row.endswith('-iii'):
    row = row.rstrip('-iii')
    row = row[0:3]+'-3'
    return row
  elif row.endswith('-iv'):
    row = row.rstrip('-iv')
    row = row[0:3]+'-4'
    return row
  elif row.endswith('-v'):
    row = row.rstrip('-v')
    row = row[0:3]+'-5'
    return row
  elif row.endswith('-vi'):
    row = row.rstrip('-vi')
    row = row[0:3]+'-6'
    return row
  elif row.endswith('-vii'):
    row = row.rstrip('-vii')
    row = row[0:3]+'-7'
    return row
  elif row.endswith('-viii'):
    row = row.rstrip('-viii')
    row = row[0:3]+'-8'
    return row
  elif row.endswith('-ix'):
    row = row.rstrip('-ix')
    row = row[0:3]+'-9'
    return row

pokedata['Generation'] = pokedata['Generation'].apply(numerical)
```

Converting Height in decimetres to metres

```python
pokedata['Height'] = pokedata['Height']/10
```

Converting Weight in Hectograms to Kilograms

```python
pokedata['Weight'] = pokedata['Weight']/10
```

```python
pokedata.rename(columns={'Height':'Height(m)','Weight':'Weight{kg}'},inplace=True)
```

Feature Engineering

```python
pokedata['Is_Pseudo_Legendary'] =  (pokedata['HP']+pokedata['Attack']+pokedata['Defense']+pokedata['Sp.Atk']+pokedata['Sp.Def']+pokedata['Speed']==600)
```

```python
pokedata['Total_Stats'] = pokedata['HP']+pokedata['Attack']+pokedata['Defense']+pokedata['Sp.Atk']+pokedata['Sp.Def']+pokedata['Speed']
pokedata['Total_Stats']
```
